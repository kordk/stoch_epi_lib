```python
fig,ax = plt.subplots(figsize = (15,7))
ax.set_facecolor('#e7e7ee')
ax.step(realization1['T'],np.array(realization1["SG"])[:,0],label = realization1["GeneIDs"][0],color = '#1f34d4',linewidth = 2)
ax.step(realization1['T'],np.array(realization1["SG"])[:,1],label = realization1["GeneIDs"][1],color = '#ee7a05',linewidth = 2)
ax.legend()
```

```python
fig,ax = plt.subplots(2,1,figsize = (15,7),tight_layout = True)
ax[0].set_facecolor('#efeff0')
ax[1].set_facecolor('#efeff0')

ax[0].plot(realization1['T'],np.array(realization1["G"])[:,0],label = realization1["GeneIDs"][0],color = '#1f34d4',linewidth = 4)
ax[1].plot(realization1['T'],np.array(realization1["G"])[:,1],label = realization1["GeneIDs"][1],color = '#ee7a05',linewidth = 4)
ax[0].step(realization1['T'],np.array(realization1["SG"])[:,0],color = '#1f34d4',linewidth = 1)
ax[1].step(realization1['T'],np.array(realization1["SG"])[:,1],color = '#ee7a05',linewidth = 1)
# ax[0].legend(fontsize = 20)
# ax[1].legend(fontsize = 20)
ax[0].set_title(realization1["GeneIDs"][0],fontsize = 20)
ax[0].set_xlabel("Time",fontsize = 18)
ax[0].set_ylabel("Transcript",fontsize = 18)
ax[0].tick_params(labelsize=14)

ax[1].set_title(realization1["GeneIDs"][1],fontsize = 20)
ax[1].set_xlabel("Time",fontsize = 18)
ax[1].set_ylabel("Transcript",fontsize = 18)
ax[1].tick_params(labelsize=14)

plt.savefig("realization1")
```
